## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
mouse_data= pd.merge(mouse_metadata, study_results, on='Mouse ID', how="left")

mouse_data.head()
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

tumor_df = mouse_data.groupby(['Drug Regimen','Timepoint']).agg(\
                                                               Mean=("Tumor Volume (mm3)", np.mean), \
                                                                Median=("Tumor Volume (mm3)", np.median), \
                                                                Variance=("Tumor Volume (mm3)", np.var), \
                                                                Std=("Tumor Volume (mm3)", np.std), \
                                                                SEM=("Tumor Volume (mm3)", st.sem)
                                                               ).round(4)
tumor_df
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 


#slicing by the drug regiments given
top_drugs= ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

top_data= mouse_data[mouse_data['Drug Regimen'].isin(top_drugs)]

#using a lambda function, i can pull the last tumor size per mouse without too much trouble
top_tumor= top_data.groupby(['Drug Regimen','Mouse ID']).agg(\
                                                             Last_tumor_size=('Tumor Volume (mm3)',lambda x: x.iloc[-1])).round(3)

#separating the final tumor size by drug to analyze volumes by drug
volume_by_drug = top_tumor.stack(level=0).unstack(level=0)

#initializing counter and running iqr calculations for each of the 4 drugs
counter=0
for drug in top_drugs:
    quartiles = volume_by_drug[drug].quantile([.25,.5,.75]).round(2)
    lowerq = quartiles[0.25].round(2)
    upperq = quartiles[0.75].round(2)
    iqr = round(upperq-lowerq,2)
    lower_bound = round(lowerq - (1.5*iqr),2)
    upper_bound = round(upperq + (1.5*iqr),2)
    
    
    if counter == 0:
        print(f"##############################################")
    print(f"{drug} IQR data is:")
    print(f"Q1 of {drug} is: {lowerq}")
    print(f"Q3 of {drug} is: {upperq}")
    print(f"IQR of {drug} is: {iqr}")
    print(f"Median of {drug} is: {quartiles[0.5]} ")
    print(f"Values below {lower_bound} for {drug} could be outliers.")
    print(f"Values above {upper_bound} for {drug} could be outliers.")
    print(f"##############################################")
    counter += 1
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
box_list= []

#dropping null values and collecting final tumor volumes into a list
for drug in top_drugs:
    box_list.append(list(volume_by_drug[drug].dropna()))

fig1, ax = plt.subplots(figsize=(9,7))
ax.set_title('Final measured Tumor volume by Drug Regimen')
ax.set_xlabel('Drug Regimen')
ax.set_ylabel('Tumor Vol (mm3)')
ax.boxplot(box_list,sym='gD')
plt.xticks([1,2,3,4],top_drugs)
    
```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
capomulin_df = mouse_data.loc[mouse_data['Drug Regimen'] == 'Capomulin']

#I selected the first mouse that showed up in the capomulin df to use for my graph 
s185_df = capomulin_df.loc[mouse_data['Mouse ID'] == 's185' ]

plt.plot(s185_df["Timepoint"], s185_df["Tumor Volume (mm3)"])

plt.title("Tumor Size over Time")
plt.ylabel('Tumor Volume (mm3)')
#plt.xlabel("TimePoints")
#kept getting TypeError: 'str' object is not callable when trying to run plt.xlabel("TimePoints")
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
capomulin_stat_df = capomulin_df.groupby(['Mouse ID']).agg(\
         
        Mouse_weight=('Weight (g)', np.mean),\
        
        Tumor_vol_mean=('Tumor Volume (mm3)', np.mean)\
                                                           
).round(4)

#plot time
scatter= plt.scatter(
    capomulin_stat_df['Mouse_weight'],
    capomulin_stat_df['Tumor_vol_mean'],
    marker='^',
    facecolors='blue',
    edgecolors='black',
    s=capomulin_stat_df['Tumor_vol_mean'],
    alpha=.75)

plt.title("Mouse weight vs. Avg. Tumor Volume")
plt.ylabel("Tumor Volume (mm3)")
#plt.xlabel("Mouse weight (g)")

#same problem here with xlabel as on the boxplot
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
x = capomulin_stat_df["Mouse_weight"]
y = capomulin_stat_df['Tumor_vol_mean']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
regress_values = x * slope + intercept
line_eq = f'y = {str(round(slope,2))}x + {str(round(intercept,2))}'
plt.scatter(x,y)
plt.plot(x,regress_values,"r-")
plt.annotate(line_eq,(17,37),fontsize=15,color="black")
plt.title("Mouse weight vs. Avg. Tumor Volume")
#plt.xlabel("Mouse weight (g)")
plt.ylabel("Tumor Volume (mm3)")

print(f'"r" value: {rvalue}')
print(f'Regression Line: {line_eq}')

#nonstop problems with plt.xlabel()
```
